```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

# Organic Product Sales per Department

The graph below displays the amount of 'organic' labeled items sold per department. The high amount of organic products purchased in [department 4] tells us that consumers value organic products in [department 4]. 

```python
df = pd.read_csv('Data/10000_transactions.csv', engine="python")

df['product_name'] = df['product_name'].str.lower()
filtered = df[df['product_name'].str.contains("organic")]
organicDeptTot = []

for depts in range (1, 21):
    tempDept = filtered[filtered['department_id'] == depts]
    totalCount = len(tempDept)
    organicDeptTot.append(totalCount)
    
barLab = []
for lab in range (1, 21):
    barLab.append(lab)

x_pos = list(range(len(barLab)))
plt.figure(figsize=[14,5])

plt.bar(x_pos,
       organicDeptTot,
       align = 'center',
       color = 'b',
       alpha = 0.5)

plt.grid()


plt.ylabel('Number of Organic Products Sold')
plt.xlabel('Department #')
plt.xticks(x_pos, barLab)
plt.title('Organic Product Sales per Department')

plt.show()

```
